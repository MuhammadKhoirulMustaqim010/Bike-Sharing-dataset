# Cleaning Data

```python
# Mengubah tipe data object menjadi datetime
datetime_colums = ["dteday"]

for column in datetime_colums:
  hour_dataset[column] = pd.to_datetime(hour_dataset[column])
  day_dataset[column] = pd.to_datetime(day_dataset[column])
```

# Visualisasi Data

```python
cnt_data = day_dataset.groupby('mnth')['cnt'].sum().reset_index()
```

```python
# Membuat visualisasi
plt.figure(figsize=(10, 6))
sns.barplot(x='mnth', y='cnt', data=cnt_data, palette='viridis')
plt.title('Total Sewa Sepeda per Bulan')
plt.xlabel('Bulan')
plt.ylabel('Total Sewa Sepeda')
plt.xticks(ticks=range(12), labels=['Jan', 'Feb', 'Mar', 'Apr', 'May', 'Jun', 'Jul', 'Aug', 'Sep', 'Oct', 'Nov', 'Dec'])
plt.show()
```

```python
seasonal_data = day_dataset.groupby('season')['cnt'].sum().reset_index()
```

```python
plt.figure(figsize=(8, 5))
sns.barplot(x='season', y='cnt', data=seasonal_data, palette='coolwarm')
plt.title('Total Sewa Sepeda Berdasarkan Musim')
plt.xlabel('Musim')
plt.ylabel('Total Sewa Sepeda')
plt.xticks(ticks=range(4), labels=['Spring', 'Summer', 'Fall', 'Winter'])
plt.show()
```

```python
weekday_data = day_dataset.groupby('weekday')['cnt'].mean().reset_index()
```

```python
plt.figure(figsize=(8, 5))
sns.barplot(x='weekday', y='cnt', data=weekday_data, palette='magma')
plt.title('Rata-rata Sewa Sepeda Berdasarkan Hari dalam Seminggu')
plt.xlabel('Hari dalam Seminggu')
plt.ylabel('Rata-rata Sewa Sepeda')
plt.xticks(ticks=range(7), labels=['Minggu', 'Senin', 'Selasa', 'Rabu', 'Kamis', 'Jumat', 'Sabtu'])
plt.show()
```

```python
plt.figure(figsize=(12, 6))
sns.lineplot(x='dteday', y='cnt', data=day_dataset)
plt.title('Tren Sewa Sepeda Sepanjang Tahun')
plt.xlabel('Tanggal')
plt.ylabel('Jumlah Sewa Sepeda')
plt.show()
```
